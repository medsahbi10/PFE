```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained('EleutherAI/gpt-neox-20b')

model = AutoModelForCausalLM.from_pretrained(
  'mosaicml/mpt-7b-chat',
  torch_dtype=torch.bfloat16,
  device_map="auto",
  trust_remote_code=True,
)
```

```python
pipe = pipeline('text-generation', model=model, tokenizer=tokenizer, max_new_tokens=70)
```

```python
pip install langchain-community
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.llms import HuggingFacePipeline
```

```python
local_llm = HuggingFacePipeline(pipeline=pipe)
```

```python
prompt_template = PromptTemplate(
    input_variables=["title", "user_story", "business_rules"],
    template=(
        "You are a test automation expert. "
        "Generate Cucumber (Gherkin) test scenarios to validate the following contract creation rules.\n\n"
        "### Title: {title}\n"
        "### User Story:\n"
        "{user_story}\n\n"
        "### Business Rules:\n"
        "{business_rules}\n\n"
        "Write clear and detailed test scenarios using the Gherkin syntax."
    )
)
```

```python
chain = LLMChain(llm=local_llm, prompt=prompt_template)
```

```python
print(chain.run({
    "title": "Contract Creation",
    "user_story": "As a Voice API administrator, I want to be able to open a contract for a client, so that they can use the API.",
    "business_rules": """
    1. Declared phone numbers (allocatedPhoneNumbers, ...) must follow the format +CCCNNNNNNNNNN, where C is a 1 to 3-digit country code and N can be up to 14 digits. The '+' character is mandatory.
    """
}))
```

```python
from langchain.prompts.few_shot import FewShotPromptTemplate
```

```python
examples = [
    {
        "Title": "Contract Creation",
        "User Story": "As a Voice API administrator, I want to be able to open a contract for a client, so that they can use the API.",
        "Business Rules": [
            "Declared phone numbers (allocatedPhoneNumbers, ...) must follow the format +CCCNNNNNNNNNN, where C is a 1 to 3-digit country code and N can be up to 14 digits. The '+' character is mandatory."
        ],
        "Test Scenarios": [
            "Scenario: Contract creation with a valid phone number\n  Given an administrator provides a phone number in the format +CCCNNNNNNNNNN\n  When they submit the contract creation request\n  Then the contract is successfully created",
            "Scenario: Contract creation with an invalid phone number\n  Given an administrator provides a phone number in an incorrect format\n  When they submit the contract creation request\n  Then an error message is displayed"
        ]
    },
    {
        "Title": "Contract Modification by Administrator",
        "User Story": "As a Voice API administrator, I want to be able to modify a contract, so that it can be adapted to client changes.",
        "Business Rules": [
            "Modifications must be logged with a timestamp.",
            "Only administrators can modify a contract."
        ],
        "Test Scenarios": [
            "Scenario: Successful contract modification\n  Given an administrator is logged into the system\n  And they provide valid modification details\n  When they submit the modification request\n  Then the contract is updated successfully\n  And the modification is logged with a timestamp",
            "Scenario: Unauthorized contract modification\n  Given a non-administrator user attempts to modify a contract\n  When they submit the modification request\n  Then the system denies the request\n  And an error message is displayed"
        ]
    },

    {
        "Title": "Contract Listing",
        "User Story": "As a Voice API administrator, I want to be able to list all contracts, so that I can manage them.",
        "Business Rules": [
            "The list must include contract IDs, client names, and statuses.",
            "Administrators must be able to filter the list by status."
        ],
        "Test Scenarios": [
            "Scenario: Successful contract listing\n  Given an administrator is logged into the system\n  When they request the list of contracts\n  Then the system displays all contracts with their IDs, client names, and statuses",
            "Scenario: Filter contracts by status\n  Given an administrator is logged into the system\n  And they apply a filter for active contracts\n  When they request the list of contracts\n  Then the system displays only active contracts"
        ]
    }
]
```

```python
example_prompt = PromptTemplate(
    input_variables=["Title", "User_Story", "Business_Rules", "Test_Scenarios"],
    template=(
        "### Title: {Title}\n"
        "### User Story:\n"
        "{User_Story}\n\n"
        "### Business Rules:\n"
        "{Business_Rules}\n\n"
        "### Test Scenarios:\n"
        "{Test_Scenarios}\n"
    )
)
```

```python
formatted_prompt = example_prompt.format(
    Title=examples[0]["Title"],
    User_Story=examples[0]["User Story"],
    Business_Rules="\n".join(f"{i+1}. {rule}" for i, rule in enumerate(examples[0]["Business Rules"])),
    Test_Scenarios="\n".join(examples[0]["Test Scenarios"])
)
```

```python
print(formatted_prompt)
```

```python
formatted_examples = [
    {
        "Title": example["Title"],
        "User_Story": example["User Story"],
        "Business_Rules": "\n".join(f"{i+1}. {rule}" for i, rule in enumerate(example["Business Rules"])),
        "Test_Scenarios": "\n".join(example["Test Scenarios"])
    }
    for example in examples
]
```

```python
few_shot_prompt = FewShotPromptTemplate(
    examples=formatted_examples,
    example_prompt=example_prompt,
    suffix=(
        "### Title: {title}\n"
        "### User Story:\n"
        "{user_story}\n\n"
        "### Business Rules:\n"
        "{business_rules}\n\n"
        "Write clear and detailed test scenarios using the Gherkin syntax."
    ),
    input_variables=["title", "user_story", "business_rules"]
)
```

```python
few_shot_prompt = FewShotPromptTemplate(
    examples=formatted_examples,
    example_prompt=example_prompt,
    suffix=(
        "### Title: {title}\n"
        "### User Story:\n"
        "{user_story}\n\n"
        "### Business Rules:\n"
        "{business_rules}\n\n"
        "Write clear and detailed test scenarios using the Gherkin syntax."
    ),
    input_variables=["title", "user_story", "business_rules"]
)
```

```python
formatted_prompt = few_shot_prompt.format(
    title="Self Contract Modification",
    user_story="As a client application of the API, I want to be able to update certain contract data that is under my control.",
    business_rules="1. Only specific fields marked as 'modifiable' can be updated.\n2. Changes must be validated before being saved."
)
```

```python
print(formatted_prompt)
```

```python
chain = LLMChain(llm=local_llm, prompt=few_shot_prompt)
```

```python
%time
result = chain.run({
    "title": "Contract Creation",
    "user_story": "As a Voice API administrator, I want to be able to open a contract for a client, so that they can use the API.",
    "business_rules": """
    1.  Declared phone numbers (allocatedPhoneNumbers, ...) must follow the format +CCCNNNNNNNNNN, where C is a 1 to 3-digit country code and N can be up to 14 digits. The '+' character is mandatory..
    """
})
print(result)
```
